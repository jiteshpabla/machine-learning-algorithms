```
import random
import math
```

```
def assign_datapoints_to_clusters(X, centroids):
  cluster_list = [] #this will have the index of the closest centroid for each data point
  objective_function = 0
  for i, point in enumerate(X):
    # euclidean distance sqrt((x1 - x2)^2 + (y1 - y2)^2)
    min_centroid_dist = math.inf
    min_centroid_index = -1
    for j, centroid in enumerate(centroids):
      current_centroid_dist = math.sqrt((point[0] - centroid[0]) ** 2 + (point[1] - centroid[1]) ** 2)
      if current_centroid_dist < min_centroid_dist:
        min_centroid_dist = current_centroid_dist
        min_centroid_index = j
    objective_function += min_centroid_dist
    cluster_list.append(min_centroid_index)
  return cluster_list, objective_function
```

```
#computing the actual centroids of the assigned clustering
def update_centroids(num_of_centroids, num_of_points, X, clusters):
  sum_centroids = np.zeros((num_of_centroids,2))
  div_centroids = np.zeros((num_of_centroids,2))
  for i, point in enumerate(X):
    sum_centroids[clusters[i]] = np.add(np.array(point), sum_centroids[clusters[i]])
    div_centroids[clusters[i]] += 1
  #print(sum_centroids)
  #print(div_centroids)
  updated_centroids = np.divide(sum_centroids, div_centroids)
  return updated_centroids
```
